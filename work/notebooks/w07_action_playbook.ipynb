## Cost vs Value

High-value pages should be reviewed first because improving them has a greater potential business impact.

Refreshing low-traffic pages may consume editorial time without producing meaningful benefits.

The playbook helps prioritize limited resources rather than replacing editorial judgment.

## Archetype → Action

| Page Type | Recommended Action |
|-----------|--------------------|
| High traffic + stale | Refresh immediately |
| High CTR + recent | Leave unchanged |
| Low CTR + good ranking | Improve title and meta description |
| Low traffic + old | Manual review before investing effort |

## Decay / Refresh Insight

Older content often shows declining engagement over time.

Pages with strong historical traffic but high staleness may benefit most from a content refresh.

This observation is based on historical data and should be interpreted as a directional trend rather than a guarantee.
